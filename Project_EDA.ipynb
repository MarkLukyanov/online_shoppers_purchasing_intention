## Data Description

* `Administrative`, `Administrative_Duration`, `Informational`, `Informational_Duration`, `Product_Related`, `Product_Related_Duration` - number of different types of pages visited by the visitor and total time spent in each of these page categories
* `BounceRates` - percentage of visitors who enter the site from that page and then leave ("bounce") without triggering any other requests to the analytics server during that session
* `ExitRates` - all pageviews to the page, the percentage that were the last in the session
* `PageValues` - the average value for a web page that a user visited before completing an e-commerce transaction
* `SpecialDay` - the closeness of the site visiting time to a specific special day (e.g. Mothers Day, Valentine's Day) in which the sessions are more likely to be finalized with transaction
* `Month` - month of the year
* `OperatingSystems` - OS of the user
* `Browser` - type of the user's browser
* `Region` - type of the user's region
* `TrafficType` - traffic type
* `VisitorType` - type of the visitor(returning or new)
* `Weekend` - whether the date of the visit is weekend
* `Revenue` - whether the user ended up with shopping or not

### Correlations

Let's look at the linear relation. Pearson correlation coefficient will help us with that. However, there will be no information about `Month` and `VisitorType`, because their values are not numerical.

```python
corr = df.corr() # correlation matrix
plt.figure(figsize=(15,15))
sns.heatmap(corr, annot=True)
plt.show()
```

We can see that there are some values, such as `ExitRates` and `PageValues` that are highly correlated with the target value - `Revenue`. This means that we will be able to make accurate predictions, using machine learning.

Now we can take a look at the more precise relation, using Spearman's rank correlation coefficient.

```python
corr = df.corr(method='spearman')
plt.figure(figsize=(15,15))
sns.heatmap(corr, annot=True)
plt.show()
```

From here we can see that the values are even more correlated.

```python
import association_metrics as am

df_cat = df.apply(lambda x: x.astype("category") if x.dtype == "object" else x)

cramersv = am.CramersV(df_cat)

cramersv.fit()
```

They correlate high enough!

The last thing to do in this part is to measure the correlation between categorical and numerical features. We'll do that using ANOVA(Analysis of Variance).

```python
from scipy.stats import f_oneway

Data = []

for c1 in df.columns:
    for c2 in df.columns:
        if df[c1].dtype == 'object' and df[c2].dtype != 'object':
            CategoryGroupLists = df.groupby(c1)[c2].apply(list)
            AnovaResults = f_oneway(*CategoryGroupLists)

            if AnovaResults[1] >= 0.05:
                Data.append({'Category' : c1, 'Numerical' : c2, 'Is correlated' : 'No'})
            else:
                Data.append({'Category' : c1, 'Numerical' : c2, 'Is correlated' : 'Yes'})

AnovaRes = pd.DataFrame.from_dict(Data)
AnovaRes
```

### Polars

Now we will compare the work rate of 2 libraries: "Pandas" and "Polars". However, the dataframe is very little and each of them will work in split seconds. That is why we need to enlarge the dataframe up to about 1_000_000 rows.

```python
!pip install polars

import polars as pl
```

```python
df_big = df.copy()

for i in range(100):
  df_big = pd.concat([df_big, df])

df_big.to_csv("customers_big.csv", index=False)
```

```python
df_big.shape, df.shape
```

1)Data Reading

```python
%%time

df_pl = pl.read_csv("customers_big.csv")
```

```python
%%time

df = pd.read_csv("customers_big.csv")
```

Data reading is about two times faster in Polars

2) Data Filtering

```python
%%time

df[(df['Administrative'] < 8) & (df['Browser'] == 2)]
```

```python
%%time

df_pl.filter((pl.col('Administrative') < 8) & (pl.col('Browser') == 2))
```

Data Filtering is about two times faster in Pandas

3) Data Aggregation

```python
%%time

df.groupby('Browser').agg({'Administrative_Duration' : 'mean', 'ExitRates' : 'max'})
```

```python
%%time

df_pl.groupby('Browser').agg([pl.mean('Administrative_Duration'), pl.max('ExitRates')])
```

Data aggregation time is almost the same

As can be seen from above, we cannot determine the fastest of these 2 libraries, because in some cases "Polars" performs faster, in the other - "Pandas".
